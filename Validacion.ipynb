# Inicio código

```python
ruta = 'D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/2022-10-01/'

rutaSensor1 = ruta + 'sensor1.csv'
rutaSensor2 = ruta + 'sensor2.csv'
rutaSensor3 = ruta + 'sensor3.csv'
rutaSensor4 = ruta + 'sensor4.csv'
rutaSensor5 = ruta + 'sensor5.csv'
rutaSensor6 = ruta + 'sensor6.csv'
rutaSensor7 = ruta + 'sensor7.csv'
rutaSensor8 = ruta + 'sensor8.csv'
rutaPluviometro = ruta + 'pluviometro.csv'

sensor1 = pd.read_csv(rutaSensor1)
sensor2 = pd.read_csv(rutaSensor2)
sensor3 = pd.read_csv(rutaSensor3)
sensor4 = pd.read_csv(rutaSensor4)
sensor5 = pd.read_csv(rutaSensor5)
sensor6 = pd.read_csv(rutaSensor6)
sensor7 = pd.read_csv(rutaSensor7)
sensor8 = pd.read_csv(rutaSensor8)
pluviometro = pd.read_csv(rutaPluviometro)

sensores = [sensor1, sensor2, sensor3, sensor4, sensor5, sensor6, sensor7, sensor8]

datosSensores = pd.concat(sensores, keys =['1', '2', '3', '4', '5', '6', '7', '8'])

datosSensores['V_t_ext'] = 0
datosSensores['V_h_ext'] = 0
datosSensores['V_t_C_cal'] = 0
datosSensores['V_h_C_cal'] = 0
datosSensores['V_t_L_cal'] = 0
datosSensores['V_h_L_cal'] = 0
# datosSensores['V_h_C'] = 0
# datosSensores['V_h_L'] = 0
```

```python
# 0 - sin comprobar
# 1 - no válidos
# 2 - válidos



datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_t_ext'] = 1
datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_h_ext'] = 1
datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_t_C_cal'] = 1
datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_h_C_cal'] = 1
datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_t_L_cal'] = 1
datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_h_C_cal'] = 1
# datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_h_C'] = 1
# datosSensores.loc[datosSensores['bateria'] < 3.3, 'V_h_L'] = 1

datosSensores.loc[(datosSensores['t_ext'].isnull()), ['V_t_ext']] = 1
datosSensores.loc[(datosSensores['h_ext'].isnull()), ['V_h_ext']] = 1
datosSensores.loc[((datosSensores['t_C_cal'].isnull()) | (abs(datosSensores['t_L_cal'] - datosSensores['t_C_cal']) > 10)), ['V_t_C_cal']] = 1
datosSensores.loc[((datosSensores['h_C_cal'].isnull()) | (datosSensores['h_C_cal'] >= 100)), ['V_h_C_cal']] = 1
datosSensores.loc[((datosSensores['t_L_cal'].isnull()) | (abs(datosSensores['t_L_cal'] - datosSensores['t_C_cal']) > 10)), ['V_t_L_cal']] = 1
datosSensores.loc[((datosSensores['h_L_cal'].isnull()) | (datosSensores['h_L_cal'] >= 100)), ['V_h_L_cal']] = 1
# datosSensores.loc[((datosSensores['h_C'].isnull()) | (datosSensores['h_C'] > 1000)), ['V_h_C']] = 1
# datosSensores.loc[((datosSensores['h_L'].isnull()) | (datosSensores['h_L'] > 1000)), ['V_h_L']] = 1

datosSensores
```

```python
datosSensores.loc[datosSensores['h_L_cal'] > 100, ['h_L_cal']] = 100
datosSensores.loc[datosSensores['h_C_cal'] > 100, ['h_C_cal']] = 100
```

# ↓ Para validar con similitud

```python
timeStamps = sensor1['ts'].to_list()
for timeStamp in timeStamps:
    # Para validat t_ext
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['t_ext'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 900000]['t_ext'].mean(skipna=True)) > 7), 'V_t_ext'] = 1

    # Para validar h_ext
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['h_ext'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 900000]['h_ext'].mean(skipna=True)) > 15), 'V_h_ext'] = 1

    # Para validar t_C_cal
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['t_C_cal'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 1800000]['t_C_cal'].mean(skipna=True)) > 10), 'V_t_C_cal'] = 1

    # Para validar t_L_cal
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['t_L_cal'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 1800000]['t_L_cal'].mean(skipna=True)) > 10), 'V_t_L_cal'] = 1

    # Para validar h_C_cal
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['h_C_cal'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 1800000]['h_C_cal'].mean(skipna=True)) > 30), 'V_h_C_cal'] = 1

    # Para calidar h_L_cal
    datosSensores.loc[(abs(datosSensores['ts'] - timeStamp) < 300000) & (abs(datosSensores['h_L_cal'] - datosSensores.loc[abs(datosSensores['ts'] - timeStamp) < 1800000]['h_L_cal'].mean(skipna=True)) > 30), 'V_h_L_cal'] = 1
```
