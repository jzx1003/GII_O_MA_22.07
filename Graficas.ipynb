## Gráficas a lo largo del tiempo de las variables más importantes

```python
datosSensores1 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor1.csv',index_col=[0])
datosSensores2 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor2.csv',index_col=[0])
datosSensores3 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor3.csv',index_col=[0])
datosSensores4 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor4.csv',index_col=[0])
datosSensores5 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor5.csv',index_col=[0])
datosSensores6 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor6.csv',index_col=[0])
datosSensores7 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor7.csv',index_col=[0])
datosSensores8 = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/Datos_Validados_sensor8.csv',index_col=[0])

datosSensores = [datosSensores1, datosSensores2, datosSensores3, datosSensores4, datosSensores5, datosSensores6, datosSensores7, datosSensores8]

datosSensores = pd.concat(datosSensores, keys=[1,2,3,4,5,6,7,8])
datosSensores
```

```python
for i in range(8):

    figure, axis = plt.subplots(6)
    figure.suptitle(f'Gráficas del sensor {str(i+1)}')
    figure.subplots_adjust(hspace=1)
    figure.set_figheight(6)
    axis[0].set_title('t_ext')
    axis[0].get_xaxis().set_visible(False)
    axis[0].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_t_ext'])
    axis[1].set_title('h_ext')
    axis[1].get_xaxis().set_visible(False)
    axis[1].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_h_ext'])
    axis[2].set_title('t_C_cal')
    axis[2].get_xaxis().set_visible(False)
    axis[2].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_t_C_cal'])
    axis[3].set_title('t_L_cal')
    axis[3].get_xaxis().set_visible(False)
    axis[3].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_t_L_cal'])
    axis[4].set_title('h_C_cal')
    axis[4].get_xaxis().set_visible(False)
    axis[4].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_h_C_cal'])
    axis[5].set_title('h_L_cal')
    axis[5].get_xaxis().set_visible(False)
    axis[5].plot(datosSensores.loc[(i+1)]['ts'], datosSensores.loc[(i+1)]['V_h_L_cal'])

    figure.supxlabel('TimeStamp')
    figure.supylabel('Validez de los datos')
    props = dict(boxstyle='round', facecolor='gainsboro', alpha=0.5)
    figure.text(0,0,s='1 - No válido\n2 - Válido', bbox=props)
```
