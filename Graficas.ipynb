## Gráficas a lo largo del tiempo de las variables más importantes

```python
datosSensores = pd.read_csv('D:/uni/4º año/TFG-GitHub/GII_O_MA_22.07/DatosValidados/SoloDatosValidos.csv',index_col=[0,1])
```
